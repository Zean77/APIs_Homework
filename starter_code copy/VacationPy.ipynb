### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#call the g_key
gmaps.configure(api_key=g_key)
```

```python
# create a heatmap that displays the humidity for every city
locations = city_df[['Lat','Lng']]
humidity = city_df['Humidity']

fig = gmaps.figure()

heat_map = gmaps.heatmap_layer(locations, weights = humidity,
                              dissipating=False, max_intensity = 10,
                              point_radius = 1.5)
fig.add_layer(heat_map)

fig
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
#https://maps.googleapis.com/maps/api/place/nearbysearch/json?
#location=-33.8670522,151.1957362&radius=500&types=food&name=harbour&key=YOUR_API_KEY

# add a 'Hotel Name' column
hotel_df = city_df.loc[:, ['City','Country', 'Lat','Lng']]
hotel_df['Hotel Name'] = ""
hotel_df.head()
```

```python
# call APIS and search all hotel information based on provided information
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

city_location = f"{hotel_df['Lat'][0]},{hotel_df['Lng'][0]}"
city_radius = 5000
search_types = 'hotel'
target_search = 'hotel'

params = {
    'location':city_location,
    'radius': city_radius,
    'type': search_types,
    'keyword': target_search,
    'key': g_key
}

response = requests.get(base_url, params=params)
pprint(response.json())
```

```python
# create a hotel_list to hold all hotels from search
hotel_list =[]

# define the params
params = {
    'radius': city_radius,
    'type': search_types,
    'keyword': target_search,
    'key': g_key
}

# loop through all cities and find nearby hotels and print all of their names
for index, column in hotel_df.iterrows():
    lat = column['Lat']
    lng = column['Lng']
    params['location'] = f'{lat}, {lng}'
    
    response = requests.get(base_url, params=params).json()
    #print(json.dumps(target_hotel, indent=4, sort_keys=True))
    try:
        hotel_list.append(response['results'][0]['name'])
    except:
        hotel_list.append(np.nan)
print(hotel_list)
```

```python
# put all hotel names into the column 'Hotel Name'
hotel_df['Hotel Name']= hotel_list
hotel_df.dropna()
hotel_df.to_csv('hotel_export.csv')
```
